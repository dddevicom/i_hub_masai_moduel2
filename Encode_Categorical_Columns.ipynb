```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})
```

```python
df
```

```python
df['age'] = df['age'].fillna(df['age'].median())
```

```python
mode_val = df['city'].mode()
df['city'] = df['city'].fillna(mode_val[0])
```

```python
df
```

```python
df.duplicated()
```

```python
#df.drop_duplicates(subset=['customer_id'], keep='last')
df = df.drop_duplicates(keep='first')
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# One-Hot Encoding on city
df = pd.get_dummies(df, columns=['city'], dtype=int)

# Label Encoding on gender
lable = LabelEncoder()
df['gender'] = lable.fit_transform(df['gender'])

print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
cols_to_scale = ['age', 'annual_income']
# scaler = MinMaxScaler()
# df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
# print("=== Min-Max Scaled ===")
# print(df)

minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[cols_to_scale] = minmax.fit_transform(df[cols_to_scale])
print("=== Min-Max Scaled ===")
print(df_minmax)
```

```python
standard = StandardScaler()
df_standard = df.copy()
df_standard[cols_to_scale] = standard.fit_transform(df[cols_to_scale])
print("=== Standard Scaled ===")
print(df_standard)
```

**MinMaxScaler vs StandardScaler — When to Prefer Each**
Use MinMaxScaler when your algorithm strictly requires inputs in a bounded range — such as Neural Networks with sigmoid/tanh activations or image pixel normalization — and you are confident your data contains no significant outliers, because a single extreme value will compress every other value toward zero and distort the entire scale.
Use StandardScaler when your data may contain outliers, or when you are using algorithms rooted in statistical assumptions like Linear Regression, Logistic Regression, SVM, or PCA — these algorithms expect features centred around zero with consistent spread, which StandardScaler directly delivers without being sensitive to extreme values.
In practice, default to StandardScaler — real-world data almost always has outliers, and StandardScaler degrades gracefully whereas MinMaxScaler degrades catastrophically, so it is the safer and more robust choice unless you have a specific architectural reason to bound your inputs
